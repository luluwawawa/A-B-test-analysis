```python
# Univariate statistics
def unistats(df):
    output_df=pd.DataFrame(columns=["count","Mising","Unique","Dtype","Numeric","Mean","Min","25%","Median","75%","Max","std","Skew","Kurt"])
    for col in categorical_col:
            output_df.loc[col]=[df[col].count(),df[col].isnull().sum(),len(df[col].unique()),df[col].dtype,pd.api.types.is_numeric_dtype(df[col]),
                                '-','-','-','-','-','-','-','-','-']
    for col in numerical_col:
        output_df.loc[col]=[df[col].count(),df[col].isnull().sum(),len(df[col].unique()),df[col].dtype,pd.api.types.is_numeric_dtype(df[col]),
                               df[col].mean(),df[col].min(),df[col].quantile(0.25),df[col].median(),df[col].quantile(0.75),df[col].max(),df[col].std(),df[col].skew(),df[col].kurt()]
    return output_df

output=unistats(df)
output.to_excel("f_univariate_statistics.xls")
```

```python
# 粉丝等级为1：
data1=new_df[(new_df["fans_level"]==0)|(new_df["fans_level"]==1)].reset_index()

# stage 1:
first_stage_1= smf.ols("stay_cnt_fans1~ ratio",data=data1).fit()
first_stage_2= smf.ols("stay_cnt_fans2~ ratio",data=data1).fit()
first_stage_3= smf.ols("stay_cnt_fans3~ ratio",data=data1).fit()
first_stage_4= smf.ols("stay_cnt_fans4~ ratio",data=data1).fit()
first_stage_5= smf.ols("stay_cnt_fans5~ ratio",data=data1).fit()
first_stage_6= smf.ols("stay_cnt_fans6~ ratio",data=data1).fit()

# stage 2:
data1=data1.assign(stay_cnt_fans1_fitted= first_stage_1.fittedvalues)
data1=data1.assign(stay_cnt_fans2_fitted= first_stage_2.fittedvalues)
data1=data1.assign(stay_cnt_fans3_fitted= first_stage_3.fittedvalues)
data1=data1.assign(stay_cnt_fans4_fitted= first_stage_4.fittedvalues)
data1=data1.assign(stay_cnt_fans5_fitted= first_stage_5.fittedvalues)
data1=data1.assign(stay_cnt_fans6_fitted= first_stage_6.fittedvalues)

second_stage= smf.ols("stay_time_s~ stay_cnt_fans1_fitted + stay_cnt_fans2_fitted + stay_cnt_fans3_fitted + stay_cnt_fans4_fitted + stay_cnt_fans5_fitted + stay_cnt_fans6_fitted",data=data1).fit()

# fist stage
intercept_coefs=[]
intercept_coefs.append(first_stage_1.params["Intercept"])
intercept_coefs.append(first_stage_2.params["Intercept"])
intercept_coefs.append(first_stage_3.params["Intercept"])
intercept_coefs.append(first_stage_4.params["Intercept"])
intercept_coefs.append(first_stage_5.params["Intercept"])
intercept_coefs.append(first_stage_6.params["Intercept"])

ratio_coefs=[]
ratio_coefs.append(first_stage_1.params["ratio"])
ratio_coefs.append(first_stage_2.params["ratio"])
ratio_coefs.append(first_stage_3.params["ratio"])
ratio_coefs.append(first_stage_4.params["ratio"])
ratio_coefs.append(first_stage_5.params["ratio"])
ratio_coefs.append(first_stage_6.params["ratio"])

intercept_pvalue=[]
intercept_pvalue.append(first_stage_1.pvalues["Intercept"])
intercept_pvalue.append(first_stage_2.pvalues["Intercept"])
intercept_pvalue.append(first_stage_3.pvalues["Intercept"])
intercept_pvalue.append(first_stage_4.pvalues["Intercept"])
intercept_pvalue.append(first_stage_5.pvalues["Intercept"])
intercept_pvalue.append(first_stage_6.pvalues["Intercept"])

ratio_pvalue=[]
ratio_pvalue.append(first_stage_1.pvalues["ratio"])
ratio_pvalue.append(first_stage_2.pvalues["ratio"])
ratio_pvalue.append(first_stage_3.pvalues["ratio"])
ratio_pvalue.append(first_stage_4.pvalues["ratio"])
ratio_pvalue.append(first_stage_5.pvalues["ratio"])
ratio_pvalue.append(first_stage_6.pvalues["ratio"])

stage1_output= pd.DataFrame(dict(
    intercept_coefs=intercept_coefs,
    ratio_coefs=ratio_coefs,
    intercept_pvalue=intercept_pvalue,
    ratio_pvalue=ratio_pvalue
))

# second stage
coefs= []
coefs.append(second_stage.params["Intercept"])
coefs.append(second_stage.params["stay_cnt_fans1_fitted"])
coefs.append(second_stage.params["stay_cnt_fans2_fitted"])
coefs.append(second_stage.params["stay_cnt_fans3_fitted"])
coefs.append(second_stage.params["stay_cnt_fans4_fitted"])
coefs.append(second_stage.params["stay_cnt_fans5_fitted"])
coefs.append(second_stage.params["stay_cnt_fans6_fitted"])

pvalue= []
pvalue.append(0)
pvalue.append(second_stage.pvalues["stay_cnt_fans1_fitted"])
pvalue.append(second_stage.pvalues["stay_cnt_fans2_fitted"])
pvalue.append(second_stage.pvalues["stay_cnt_fans3_fitted"])
pvalue.append(second_stage.pvalues["stay_cnt_fans4_fitted"])
pvalue.append(second_stage.pvalues["stay_cnt_fans5_fitted"])
pvalue.append(second_stage.pvalues["stay_cnt_fans6_fitted"])

stage2_output= pd.DataFrame(dict(
    coefs=coefs,
    pvalue=pvalue
))

print(stage1_output)
print(stage2_output)

stage1_output.to_excel("fan1_stage1.xls")
stage2_output.to_excel("fan1_stage2.xls")
```
